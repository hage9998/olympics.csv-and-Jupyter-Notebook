```python
import pandas as pd
pd.set_option('display.max_rows', 100)
```

```python
names_columns = ['Countrys', 'N°Games Summer', 'GoldS', 'SilverS', 'BronzeS', 'Total MedalsS', 'N°Games Winter', 'GoldW', 'SilverW', 'BronzeW', 'Total MedalsW', 'Total Games', 'GoldT', 'SilverT', 'BronzeT', 'Total GeralT']
df = pd.read_csv('olympics.csv', skiprows=[0, 1], names=names_columns, nrows = 100 ) #Read csv files
```

```python
df
```

```python
df.loc[:,'Countrys'] = df.loc[:,'Countrys'].str.split('(', expand = True)[0] #Change Countrys values
df.loc[:,'Countrys'] = df.loc[:,'Countrys'].map(lambda x: x[:-1]) 
df.loc[:,'Countrys']
```

```python
get_gold_summer = df.loc[df['GoldS'] == df['GoldS'].max()] #Country with more Gold medals in summer
```

```python
get_gold_winter = df.loc[df['GoldW'] == df['GoldW'].max()] #Country with more Gold medals in winter
get_gold_winter
```

```python
con = pd.concat([get_gold_summer, get_gold_winter])
con
```

```python
import matplotlib.pyplot as plt
```

```python
plot_change = plt.plot([con['GoldS'], con['GoldW']])
plt.legend(plot_change, (get_gold_summer['Countrys'].values[0], get_gold_winter['Countrys'].values[0]))
#Changes with the country with more gold medals in summer and winter
```

```python
names = df.loc[:,'Countrys']
plt.figure(figsize=(20, 3))
plt.xticks(rotation=90)
plt.bar(names, df.loc[:,'GoldS'])
plt.figure(figsize=(20, 3))
plt.xticks(rotation=90)
plt.bar(names, df.loc[:,'GoldW'])
#Gold medals per country in each season
```

```python
more_medals = df.loc[df['Total GeralT'] == df['Total GeralT'].max(), 'Countrys'].values[0]
more_goldMedals = df.loc[df['GoldT'] == df['GoldT'].max(), 'Countrys'].values[0]
{'More gold Medals' : more_goldMedals, 'More Medals in Total':more_medals}, more_medals == more_goldMedals
#Verify if the country with more medals has more gold medals too
```

```python
total_medals_summer = df.loc[:,'Total MedalsS']
total_medals_winter = df.loc[:,'Total MedalsW']
get_better = []
for i in range(len(df)):
   if (total_medals_winter[i] >= total_medals_summer[i]):
         get_better.append(df.iloc[i, 0])
get_better #Countrys that get better after first season
```

```python
total = df.loc[:,'Total MedalsS'].sum()
df['Porcentagem'] = df['Total MedalsS'].map(lambda x: x/total * 100)
df.rename(columns={'Porcentagem':'Percentage'}, inplace = True) #Rename a column
df #Add a column contain the percentage that country has in medals total amount 
```
